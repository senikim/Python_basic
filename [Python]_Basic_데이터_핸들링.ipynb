### 3) 데이터 추출하기
* 행 단위로 추출하기
    * 인덱스 : 숫자 대신 first, second, third 등의 문자열로도 지정 가능
    * 행번호 : 데이터의 순서를 따라간다. 실제 데이터프레임 값에서 확인할 수 없으며 정수만으로 데이터 조회
    * loc(), iloc()


#### (1) loc()을 이용하여 인덱스 기준으로 행 데이터 추출하기

```python
df.loc[0]
    # 0번째 인덱스 추출
df.loc[[0]]
    # 데이터 프레임 형태로 출력
```

```python
# 여러 행을 테이블 형태로 출력

df.loc[[0, 99, 999]]
```

#### (2) iloc을 사용하여 행번호 기준으로 행 데이터 추출하기

```python
df.iloc[0:9]
```

```python
df.iloc[[0, 99, 999]]
```

```python
# 마지막 행 출력

df.iloc[[-1]]
    # 1703 행
    # 마지막 행 출력을 위해 음수 입력해도 에러 X
```

#### (3) 데이터의 행과 열을 지정하여 출력
* df.loc[[행],[열]] -> 문자열 리스트로
* df.iloc[[행], [열]] -> 정수로 

```python
df.loc[:, ['year', 'pop']]
    # loc() 사용시 문자열 리스트를 전달해야 함 
```

```python
df.iloc[:, [2,4]]
    # iloc() 사용시 정수로 리스트를 전달해야 함
```

### 몸풀기 QUIZ

country, gdpPercap 열 추출을 위해 loc와 iloc을 어떻게 사용해야 할까?

```python
# 1. loc로 모든 행 데이터의 country, gdpPercap 열 추출하기
df.loc[:, ['country', 'gdpPercap']]
```

```python
# 2. iloc로 모든 행 데이터의 country, gdpPercap 열 추출하기

# 열 번호 출력
df.columns.get_loc("country")  ## 0
df.columns.get_loc("gdpPercap")  ## 5

df.iloc[:, [0, 5]]
```

```python
# 3. loc를 사용해서 인덱스가 10인 행부터 13인 행의 country, lifeExp, gdpPercap 열 데이터 추출하기
# 주의! loc는 인덱스 기준으로 행을 찾아가므로 슬라이싱을 할 때 [이상:이하]로 인식한다.

df.loc[10:13, ["country", "lifeExp", "gdpPercap"]]
```

```python
# 4. iloc를 사용해서 인덱스가 10인 행부터 13인 행의 country, lifeExp, gdpPercap 열 데이터 추출하기

# 열 번호 출력
df.columns.get_loc("lifeExp") ## 3

df.iloc[10:14, [0, 3, 5]]
    # iloc 는 행 번호 기준으로 slicing -> [n:m]이면 [n:m-1] 까지 출력됨
```

### 3) 기초통계 계산하기



#### (1) lifeExp 열을 연도별로 그룹화하여 계산하기


```python
df.groupby('year')['lifeExp'].mean()

# 데이터 프레임 형태로 출력 # [[]]

df.groupby('year')[['lifeExp']].mean()
```

#### (2) lifeExp, gdpPercap 열의 평균값을 연도, 대륙별로 그룹화하여 한 번에 계산하기

```python
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

#### (3) 그룹화한 데이터 개수 세기 (빈도)

```python
# continent로 그룹화하여 country 열만 추출해서 데이터 빈도수 계산

df.groupby('continent')['country'].nunique()
    # nunique() : 빈도수 개수
    # unique() : 겹치지 않는 것들의 변수명
```

### 1) 시리즈, 데이터 프레임 만들기

#### (2) 데이터 프레임 만들기

```python
# 딕셔너리를 dataframe 클래스에 전달

scientists = pd.DataFrame({
    'Name' : ['Rosaline Frankline', 'William Gosset'],
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-04-16', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]
})
print(scientists)
    # 인덱스 자동 생성
```

### 2) 데이터 프레임 다루기

```python
# 파일 임포트

scientists = pd.read_csv("/content/drive/MyDrive/Pandas_practice/scientists.csv")

scientists.head()
```

#### (1) 통계내기

```python
# 나이의 최솟값
print(scientists['Age'].min())
# 나이의 최댓값
print(scientists['Age'].max())
# 나이의 평균
print(scientists['Age'].mean())
```

```python
scientists.describe()
```

#### (2) 특정 조건 만족하는 행만 추출

```python
# 평균보다 나이가 어리다는 조건을 만족하는 행만 추출
scientists[scientists['Age'] < scientists['Age'].mean()]
```

### 3) 데이터 프레임의 데이터 처리하기

#### (1) 날짜를 datetime 자료형으로 바꾸기

Born과 Died 열은 현재 object.
날짜이므로 datetime 자료형으로 바꾸자.

```python
print(scientists['Born'].dtypes)
print(scientists['Died'].dtypes)
```

```python
# Born 열을 datetime 자료형으로 바꿔서 born_datetime 변수에 저장하기
# pd.to_datetime
# format = '%Y-%m-%d'

born_datetime = pd.to_datetime(scientists['Born'], format = '%Y-%m-%d')
print(born_datetime)
```

```python
died_datetime = pd.to_datetime(scientists['Died'], format = '%Y-%m-%d')
print(died_datetime)
```

#### (2) 새로운 열 추가하기

```python
# scientists 데이터 프레임에 born_dt, died_dt 추가

scientists['born_dt'] = born_datetime
scientists['died_dt'] = died_datetime
scientists.head()
```

#### (3) 시간 계산하기

died_dt 열에서 born_dt를 빼면 시간 계산 가능

```python
# 과학자들이 살아있던 날 계산 후 열에 추가

scientists['age_days_dt'] = scientists['died_dt']-scientists['born_dt']
scientists.head()
```

#### (4) 열 삭제하기

drop( , axis = , inplace = )
* axis = 0 : 행
* axis = 1 : 열
* inplace = True : 덮어씌우기

```python
# age_days_dt 열 삭제

scientists_dropped = scientists.drop(['age_days_dt'], axis = 1)
scientists_dropped
```

```python
# 원본 데이터 프레임에서 drop
# inplace = True

scientists.drop(["age_days_dt"], axis = 1, inplace = True)
scientists
```

### 4) 데이터 저장하고 불러오기

```python
# 저장하기

names = scientists['Name']
    # Name 열만 저장할 것이다 (시리즈)
```

#### (1) csv 파일로 저장하기

```python
names.to_csv('name.csv')
```

#### (2) 엑셀 파일로 저장하기

시리즈 자료형은 to_frame() 함수를 사용해 데이터프레임으로 변환 후 엑셀로 저장

```python
# 데이터 프레임으로 변환
names_df = names.to_frame()

# xlsx 확장자로 저장
import openpyxl
names_df.to_excel('names.xlsx')
```
